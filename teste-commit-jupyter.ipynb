## O que Ã© LÃ³gica de programaÃ§Ã£o?
Antes de escrever cÃ³digo em uma linguagem de programaÃ§Ã£o, Ã© fundamental entender a lÃ³gica por trÃ¡s da soluÃ§Ã£o de um problema. Um algorÃ­timo Ã© uma sequÃªncia de passos finitos e bem definidos para resolver um problema. O pseudocÃ³digo pe uma forma de representar esses passos de maneira informal e prÃ³ximma da linguagem humana. Antes de traduzi-la para Python.

## o que Ã© PseudocÃ³digo?
PseudocÃ³digo Ã© uma forma de descrever algorÃ­timos usando uma linguagem intermediÃ¡ria entre o portuguÃªs ( ou outro idioma nartural) e uma linguagem de programaÃ§Ã£o real.

Ele nÃ£o sugue regras rÃ­gidas de sintaxe, mas utiliza estruturas comuns da progrmaÃ§Ã£o ( como se, entÃ£o,senÃ£o, enquanto, para cada, funÃ§Ã£o,etc.),escritas de maneira simples e legÃ­vel para humanos.

ðŸ‘‰ A ideia Ã© focar na lÃ³gica do algoritmo, sem se preocupar com os detalhes tÃ©cnicos da linguagem que serÃ¡ usada depois.

Vejamos um exemplo de PseudocÃ³digo:

INÃCIO

LEIA nota1

LEIA nota2

CALCULE media = (nota1 + nota2) / 2

ESCREVA "A mÃ©dia do aluno Ã©:", media

SE media >= 7.0 ENTÃƒO

    ESCREVA "Aluno Aprovado!"

SENÃƒO

    ESCREVA "Aluno Reprovado."
FIM

traduÃ§Ã£o para o Python:

```python
#traduzindo o pseudocÃ³digo para Python
nota1 = 7.5
nota2 = 8.0

media = (nota1 + nota2) / 2

print(f"A mÃ©dia do aluno Ã©: {media}")
if media >= 7.0:
    print("Aluno Aprovado!")
else:
    print("Aluno Reprovado")
```

## 5.1 Operadores Aritimeticos

Usados para realizar operaÃ§Ãµes matemÃ¡ticas.

```python
# DefiniÃ§Ã£o de variÃ¡veis
a = 10
b = 3
```

## 10. Estruturas de Dados - Conjuntos (Sets)
Conjuntos sÃ£o coleÃ§Ãµes nÃ£o ordenads de intens Ãºnicos e mutÃ¡veis. Eles sÃ£o Ãºteis para remover duplicats e realizar operaÃµes matemÃ¡ticas de conjuntos()uniÃ£, interseÃ§Ã£o).

```python
# Criando um conjunto (note que os intens duplicados sÃ£o removidos)
numeros = {1, 2, 3, 4, 2, 3, 5}
print(f"Conjunto de nÃºmeros (sem duplicatas): {numeros}")
```

```python
type(numeros)
```

```python
# Adicionando um item
numeros.add(6)
print(f"ApÃ³s adicionar o valor 6: {numeros}")
```

```python
# Removendo um item
numeros.remove(2)
print(f"ApÃ³s remover o 2: {numeros}")
```

```python
# OperaÃ§Ãµes de conjunto
conjunto_a = {1, 2, 3, 4}
conjunto_b = {3, 4, 5, 6}
```

```python
# UniÃ£o (todos elementos de ambos os conjuntos)
uniao = conjunto_a.union(conjunto_b)
print(f"UniÃ£o de A e B: {uniao}")
```

```python
# InterseÃ§Ã£o ( elementos que estÃ£o em ambos conjuntos)
intersecao = conjunto_a.intersection(conjunto_b)
print(f"InterseÃ§Ã£o de A e B: {intersecao}")
```
